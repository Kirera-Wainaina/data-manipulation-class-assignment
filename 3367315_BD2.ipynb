### Do user-given ratings match GPT ratings?

Users often give their reviews and scores. Both of these provide information that influence potential buyers.
If we were to give an independent entity, GPT, some of these reviews and ask it to rate based on the sentiment of the reviews.
To what degree would the user scores correlate with the GPT scores 

```python
from openai import OpenAI
```

```python
# get OPENAI api key from file
api_key = None
file_with_api_key = ".env"

# to make it easy on yourself, simply create a file with the following line
# OPENAI_API_KEY=my_api_key
# the code will pick your API key through string operations

with open(file_with_api_key) as env:
    for line in env:
        if line.startswith("OPENAI_API_KEY"):
            api_key = line.removeprefix("OPENAI_API_KEY=").strip()
            break
```

```python
# It'll be too expensive to run using the whole reviews dataframe of over 39,000 records
# We take a slice of reviews where each row has both a score and a review

# We also eliminate the reviews with a score of 0 because the user probably 
# didn't give a rating after writing the review

sample_reviews = reviews.dropna().query('Score > 0')[:100]
sample_reviews
```

```python
# We take the reviews and send them to GPT with the instruction to rank them from 1 - 5


# We create a function that takes a review and returns an integer of the score as given by GPT
def get_score_of_review(review):
    client = OpenAI(api_key=api_key)    

    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {
                "role": "system", 
                 "content": "You are required to rate the review that is \
                assigned to you on a score of 1 - 5 based on the sentiment. \
                1 being the user review sounded unhappy or disatisfied and \
                5 being a customer who is happy with the product. Please \
                return only a number without any text explaining your reasoning."
            },
            {
                "role": "user",
                "content": review
            }
        ]
    )
    return int(completion.choices[0].message.content)
```

```python
# This part of the code takes about 2 minutes to run

scores = [get_score_of_review(i) for i in sample_reviews['Review']]
```

```python
# We then enter the scores into the sample reviews dataframe
sample_reviews['gpt_scores'] = scores
```

```python
sample_reviews
```

We have now integrated the gpt_scores into the sample_reviews dataframe

```python
# Calculate the correlation coefficient between gpt scores and and user scores

sample_reviews.filter(['Score', 'gpt_scores']).corr()
```

The correlation coefficient is very close to 0 indicating that we can't use user-given scores to predict GPT scores or vice versa.

We will plot a scatter matrix where the `Score` is on the x-axis and the `gpt_scores` are on the y-axis

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(sample_reviews['Score'], sample_reviews['gpt_scores']);
ax.plot(range(0, 7, 1), range(0, 7, 1), 'r');

ax.set_title('Scatter Matrix of GPT Scores on User Scores');
ax.set_ylabel('GPT Scores');
ax.set_xlabel('User Scores');
ax.set_xticks(range(0, 7, 1));
ax.set_yticks(range(0, 7, 1));
```

#### Conclusion
- The red line in chart shows when the scores given by users match those given by GPT. Ideally, the scores should be close to this line.
- Scores that fall below the red line are scores where the user is more generous than GPT.
- Conversely, the dots above the red line are cases when GPT is more generous with the score then the user. From the sample we took it appears this is the case.
